```python
from sqlalchemy import create_engine

engine = create_engine(
    "mysql+pymysql://root:ajay@localhost/ipl_analytics"
)
```

```python
matches.to_sql(
    "matches",
    con=engine,
    if_exists="replace",
    index=False
)
```

```python
balls.to_sql(
    "balls",
    con=engine,
    if_exists="replace",
    index=False
)
```

```python
players.to_sql(
    "players",
    con=engine,
    if_exists="replace",
    index=False
)
```

```python
teams.to_sql(
    "teams",
    con=engine,
    if_exists="replace",
    index=False
)
```

```python
query = """
SELECT
match_winner,
COUNT(*) wins
FROM matches
GROUP BY match_winner
ORDER BY wins DESC
"""

result = pd.read_sql(query, engine)

print(result)
```

```python
result.to_csv(
    "team_wins.csv",
    index=False
)
```
